```python
import yfinance as yf


ticker = input("Digite o código da ação: ")
dados = yf.Ticker(ticker).history("6mo")
fechamento = dados.Close
fechamento.plot()
```

```python
# passo 03: enviar automaticamente o email com a análise
    # abrir um nova aba (ctrl + t)
    # digitar endereço do site
    # clicar no botão escrever
    # digitar o destinatario (tab)
    # digitar assunto (tab)
    # preencher o corpo do email
    #clicar no botão enviar

# !pip install pyautogui - instalar se caso nao tiver instalado

# pyautogui.click -> clique com o mouse
# pyautogui.write -> escrever um texto
# pyautogui.press -> apertar uma tecla
# pyautogui.hotkey -> apertar uma combinação de teclas (ex: Ctrl + D)
# Passo a passo

pyautogui.PAUSE = 2

pyautogui.hotkey("ctrl", "t")
pyautogui.write("https://mail.google.com/mail/u/0/#inbox")
pyautogui.hotkey("ctrl", "v")
pyautogui.press("enter")

time.sleep(5)

pyautogui.click(x=80, y=241)

pyautogui.write("marcelojonil09@gmail.com")
pyautogui.press("tab")
pyautogui.press("tab")
pyperclip.copy("Relatório de análise de mercado")
pyautogui.hotkey("ctrl", "v")
pyautogui.press("tab")

texto = f"""
Prezados,

Segue o relatório de cotação das ações da Petrobras

Cotação máxima: R${maxima:,.2f}
Cotação mínima: R${minimo:,.2f}
Cotação atual: R${atual:,.2f}

Qualquer dúvida estou a disposição

Att

Marcelo Santana
"""
pyperclip.copy(texto)
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("ctrl", "enter")
```
